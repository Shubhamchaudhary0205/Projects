```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [2, 4, 6, 8]

plt.plot(x, y)
plt.title("Simple Line Graph")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.grid(True)
plt.show()
```

```python
def chatbot():
    print("🤖 ChatBot: Hello! I’m ChatBot. Type 'bye' to exit.")

    while True:
        user = input("You: ").lower()

        if "hello" in user or "hi" in user:
            print("🤖 ChatBot: Hey there!")
        elif "how are you" in user:
            print("🤖 ChatBot: I'm just code, but I'm doing great! 😄")
        elif "name" in user:
            print("🤖 ChatBot: I'm a Python-powered chatbot.")
        elif "bye" in user:
            print("🤖 ChatBot: Goodbye! Have a great day!")
            break
        elif "help" in user:
            print("🤖 ChatBot: You can say things like 'hello', 'your name', or ask how I am.")
        else:
            print("🤖 ChatBot: Sorry, I didn't understand that.")

chatbot()
```
